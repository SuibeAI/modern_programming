# AI Coding Agent 与 Git

预计阅读时间：10 分钟。

AI Coding Agent 可以快速修改代码，但 Git 是审查和回退这些修改的安全边界。

## 原理

使用 AI Agent 时，必须保留人工审查环节：

- 修改前先看 `git status`；
- 修改后用 `git diff` 审查；
- 能运行就运行测试或最小命令；
- 小步提交，方便回滚；
- 不把 AI 生成代码直接当作正确结果。

## 推荐流程



```python
git status
git switch -c exp/agent-refactor
# 让 AI Agent 修改代码
git diff
python train.py --config config.yaml
git add train.py model.py
git commit -m "Refactor training loop"
```



关键点是分工清楚：人提供规格说明，AI 负责具体修改；人审查差异、运行验证，并决定是否提交。

## 常见风险

- AI 修改了你没有要求的文件；
- AI 删除了重要逻辑；
- AI 引入了新依赖但没有说明；
- AI 修改代码后没有运行验证。

这些风险都可以通过 `git status` 和 `git diff` 先发现。

## 练习

让 AI Agent 修改一个小函数，例如把训练日志打印得更清楚。修改后执行：



```python
git diff
```



只在你看懂修改并能运行验证后提交。
