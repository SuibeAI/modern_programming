# Git 版本管理概览

预计阅读时间：10 分钟。

Git 是记录项目历史的工具。它可以帮助我们追踪修改、回到旧版本，并在多人或 AI Agent 参与开发时审查代码变化。

## 为什么需要 Git

项目开发经常会反复修改代码、配置和文档。Git 可以帮助我们回答：

- 代码改了什么；
- 某次运行对应哪份代码；
- 改坏后如何回退；
- AI Agent 修改了哪些文件；
- 多人协作时如何审查修改。

## 示例项目

本章使用一个简化项目目录：

```text
dl-demo/
  train.py
  model.py
  config.yaml
  requirements.txt
  README.md
```

Git 适合管理代码、配置和说明文档。大型数据、模型权重和大量运行输出通常不直接放入 Git。

## 练习

判断下面哪些文件适合提交到 Git：

- `train.py`
- `config.yaml`
- `model.pth`
- `runs/exp1/events.out`
- `README.md`
- `.env`

原则：代码和小型配置可以提交；数据、权重、日志、密钥不要提交。
