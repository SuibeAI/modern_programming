## 忽略文件与实验产物

Git 不适合保存大文件和频繁变化的实验产物。深度学习项目通常需要维护 `.gitignore`。

常见不提交内容：

```text
data/
datasets/
runs/
logs/
wandb/
*.pt
*.pth
*.ckpt
__pycache__/
.ipynb_checkpoints/
.venv/
.env
```

## 合并与冲突

当两个分支修改同一段内容时，Git 可能产生冲突。

冲突标记类似：

```text
<<<<<<< HEAD
learning_rate: 0.001
=======
learning_rate: 0.0003
>>>>>>> exp/lr
```

遇到冲突时不要盲目删除标记，要读懂两边修改，再决定保留哪一版。

## worktree

`git worktree` 可以让同一个仓库同时检出多个工作目录。例如，一个目录跑 baseline，另一个目录尝试新实验。

例如当前仓库是 `dl-project`，可以在旁边创建一个新工作目录，用新分支尝试注意力模块：



````python
git worktree add ../dl-project-attention exp/attention
````



此时可以这样分工：

- `dl-project/`：保留 `main` 或 `baseline`，继续跑稳定实验；
- `dl-project-attention/`：切到 `exp/attention`，修改模型并跑新实验。

这样两个实验目录互不覆盖，但共享同一个 Git 仓库历史。它适合进阶实验管理，但会增加目录和分支理解成本，本课程只提概念。
