远程仓库用于备份代码、跨机器同步实验项目，以及和他人协作。

## Git 远程仓库

Git 是版本控制工具。远程仓库是托管在服务器上的 Git 仓库，用于备份代码、同步多台机器上的工作，以及支持多人协作。

常见远程仓库平台包括：GitHub、Gitee 和 GitLab。GitHub 是国际上最常用的平台之一；Gitee 在国内访问更方便；GitLab 常用于企业或学校自建代码平台。

课程中，远程仓库主要用于：

- 备份个人代码；
- 在多台机器之间同步实验项目；
- 提交作业或项目代码；
- 通过 pull request 进行代码审查。

## 常用命令

| 命令 | 作用 |
| --- | --- |
| `git clone` | 复制远程仓库到本地 |
| `git fetch` | 下载远端更新，并更新远程跟踪分支 |
| `git pull` | 获取远端更新，并合并到当前分支 |
| `git push` | 把本地提交推送到远端 |

## git fetch 做了什么

`git fetch` 只做两件事，但都很重要：

1. 从远程仓库下载本地缺失的提交记录和相关对象；
2. 更新远程跟踪分支，例如 `origin/main`。

它不会修改工作区，不会修改暂存区，也不会自动移动当前分支 `main`。

## 远程跟踪分支

`origin/main` 不是远程服务器上的分支本身，而是本地仓库中记录远程 `main` 分支位置的引用。

可以把它理解为一个本地只读书签：

- `origin`：远程仓库的默认名字；
- `origin/main`：本地记录的远程 `main` 分支位置；
- `git fetch` 会更新 `origin/main`；
- 本地开发通常在 `main`、`dev` 等本地分支上进行，不直接在 `origin/main` 上提交。

查看本地分支和远程跟踪分支：



````python
git branch -a
````



查看包含远程跟踪分支的提交图：



````python
git log --oneline --graph --decorate --all -10
````

## git pull

`git pull` 可以理解为：先 `git fetch`，再把远端更新合并到当前分支。



````python
git pull
````



在当前分支跟踪 `origin/main` 的常见情况下，可以近似理解为：



````python
git fetch origin
git merge origin/main
````



也就是先更新 `origin/main`，再把 `origin/main` 合并到当前分支。

## git push

`git push` 则是把本地分支上的新提交推送到远程仓库：



````python
git push
````



因此，一个常见同步流程是：



````python
git fetch origin
git log --oneline --graph --decorate --all -10
git pull
git push
````

## 示例

第一次取得远程仓库：



````python
git clone <repo-url>
cd <repo-name>
````



查看远程仓库和远程跟踪分支：



````python
git remote -v
git fetch origin
git branch -a
git log --oneline --graph --decorate --all -10
````



如果确认要把远端更新合并到当前分支，再运行：



````python
git pull
````



本地提交完成后，可以推送到远程仓库：



````python
git push
````



推送前先确认：

- 没有提交数据集和权重；
- 没有提交 `.env`、token、私钥；
- `README.md` 或实验说明已更新。



完成一次：
