# Jupyter Notebook

Jupyter Notebook 是一种把文字、代码和运行结果放在同一个文件中的交互式文档。本课程的主要内容会以 Notebook 形式呈现。

## Jupyter 的定位

普通 `.py` 文件适合编写完整程序，Notebook 更适合教学、实验和探索。

Notebook 的特点是：

- 可以分段运行代码；
- 可以在代码旁边写解释文字；
- 可以直接保存输出结果；
- 适合展示数据分析、模型实验和课堂练习。

## 基本原理

一个 `.ipynb` 文件本质上是一个 JSON 文档，里面保存了多个 cell。

常见 cell 类型：

| 类型 | 作用 |
| --- | --- |
| Markdown cell | 写说明文字、标题、公式和列表 |
| Code cell | 写代码并运行 |

运行代码时，Notebook 会连接到一个 kernel。对 Python Notebook 来说，kernel 通常就是某个 Python 环境。

## Kernel 和环境

Notebook 能否 `import` 某个包，取决于当前 kernel 使用的是哪个 Python 环境。

在 Notebook 中可以运行：

```python
import sys
print(sys.executable)
```

这能帮助你确认当前 Notebook 实际使用的 Python 解释器。

## VSCode 中的 Jupyter 插件

在 VSCode 中打开和运行 `.ipynb` 文件，需要安装 Microsoft 的 Jupyter 插件。通常还需要安装 Python 插件。

基本使用流程：

1. 安装 Python 插件和 Jupyter 插件；
2. 打开 `.ipynb` 文件；
3. 点击右上角 Select Kernel；
4. 选择当前项目对应的 Python 环境；
5. 逐个运行 code cell。

## 本课程如何使用 Notebook

本课程用 Notebook 承载主要讲义、示例代码和练习。阅读时要注意：

- Markdown cell 用来阅读概念；
- Code cell 用来运行和验证；
- 运行前先确认 kernel 是否正确；
- 如果包找不到，先检查当前 kernel 的 Python 环境。

## 常见问题

| 问题 | 可能原因 |
| --- | --- |
| 找不到包 | kernel 不是安装该包的 Python 环境 |
| 变量结果和预期不同 | cell 运行顺序混乱 |
| 修改代码后结果没变 | 对应 cell 没有重新运行 |
| Notebook 无法运行 | Jupyter 插件或 Python 插件未安装 |

遇到问题时，优先检查 kernel 和 cell 运行顺序。
