# Python 解释器、包与导入路径

Python 解释器负责运行 `.py` 文件。Python 包是可以被 `import` 的代码集合。导入路径决定 Python 去哪里查找这些包。

## 查看当前解释器

在终端中运行：

```bash
which python3
python3 --version
```

在 Python 中查看：

```bash
python3 -c "import sys; print(sys.executable)"
```

`sys.executable` 显示当前 Python 程序实际使用的解释器路径。

## Python 包是什么

包可以理解为别人写好的 Python 代码。例如：

```python
import os
import numpy
```

`os` 是 Python 标准库，通常不用额外安装；`numpy` 是第三方包，通常需要安装。

## Python 去哪里找模块

查看当前导入路径：

```bash
python3 -c "import sys; print('\n'.join(sys.path))"
```

`sys.path` 中的目录，就是 Python 执行 `import` 时会搜索的位置。

## PYTHONPATH

`PYTHONPATH` 是一个环境变量，可以额外指定 Python 的导入路径。

查看当前设置：

```bash
echo $PYTHONPATH
```

如果为空，不一定有问题。多数项目并不需要手动设置 `PYTHONPATH`。

## 临时修改导入路径

可以临时把当前目录加入导入路径：

```bash
export PYTHONPATH=$PWD:$PYTHONPATH
```

这只影响当前终端。关闭终端后设置会失效。

不要把不理解的路径长期写入 shell 配置文件，否则可能导致不同项目互相影响。

## 常见问题

`ModuleNotFoundError` 通常表示当前解释器找不到这个包。此时不要急着重复安装，先确认：

- 当前用的是哪个 `python`；
- 当前用的是哪个 `pip`；
- 包是否安装到了同一个环境。
