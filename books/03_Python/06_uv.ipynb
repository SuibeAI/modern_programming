# uv

`uv` 是较新的 Python 包和项目管理工具，特点是速度快、命令统一。

## 安装依赖

在已有项目中，可以使用：

```bash
uv pip install requests
```

如果当前没有虚拟环境，`uv` 通常会提示你先创建或指定环境。

## 创建项目

```bash
uv init demo-project
cd demo-project
uv add requests
```

`uv add` 会把依赖写入项目配置文件。

## 运行命令

```bash
uv run python --version
```

`uv run` 会在项目环境中运行命令，适合减少“包安装到了哪里”的混乱。
