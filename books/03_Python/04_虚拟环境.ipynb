# 虚拟环境

虚拟环境用于隔离不同项目的 Python 包，避免项目之间互相影响。

## 创建虚拟环境

在项目目录中运行：

```bash
python3 -m venv .venv
```

`.venv` 是常见的虚拟环境目录名。

## 激活虚拟环境

macOS / Ubuntu：

```bash
source .venv/bin/activate
```

Windows PowerShell：

```powershell
.venv\Scripts\Activate.ps1
```

激活后，终端前面通常会出现 `(.venv)`。

## 退出虚拟环境

```bash
deactivate
```

虚拟环境目录通常不提交到 Git，可以在 `.gitignore` 中加入：

```text
.venv/
```

## 遇到的问题

`venv` 可以隔离 Python 包，但它通常不负责管理 Python 本身的版本，也不擅长处理科学计算、深度学习中较复杂的二进制依赖。
