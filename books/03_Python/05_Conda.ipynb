# Conda

Conda 是常用的环境和包管理工具，适合需要科学计算、深度学习或多 Python 版本的项目。

## 创建环境

```bash
conda create -n myenv python=3.11
```

`myenv` 是环境名，可以按项目命名。

## 激活和退出

```bash
conda activate myenv
python --version
conda deactivate
```

激活后安装的包会进入当前 Conda 环境。

## 查看环境

```bash
conda env list
conda list
```

`conda env list` 查看所有环境，`conda list` 查看当前环境中的包。
