# Python 环境管理概览

本章介绍 Python 开发中最常见的环境问题：解释器、包、导入路径、依赖安装、虚拟环境、Conda、uv、Python 调试和 Jupyter Notebook。

目标不是记住所有工具细节，而是理解：代码由哪个 Python 运行、包安装到了哪里、为什么有时 `import` 会失败，以及 Notebook 使用的是哪个 Python 环境。

## 学习目标

完成本章后，你应该能够：

- 判断当前使用的是哪个 Python 解释器；
- 理解 Python 包和 `import` 的基本关系；
- 使用 `pip` 安装依赖；
- 理解虚拟环境、Conda 和 uv 的作用；
- 使用 `pdb` 和 VSCode 调试器做最基本的断点调试；
- 理解 Jupyter Notebook、kernel 和 Python 环境的关系。

## 为什么需要这些工具

`pip` 可以安装 Python 包，但它只解决“安装依赖”的问题。

当多个项目需要不同版本的依赖时，就需要 `venv` 创建独立环境，避免项目之间互相影响。

当项目还需要管理 Python 版本，或依赖包含科学计算、深度学习等较复杂的二进制包时，`Conda` 更常用。

`uv` 是较新的工具，目标是用更快、更统一的方式管理包、虚拟环境和项目依赖。

环境配置正确后，程序仍然可能有逻辑错误，这时需要使用 `pdb` 或 VSCode 调试器观察变量和执行流程。

本课程主要使用 Jupyter Notebook，因此还需要理解 Notebook 的 kernel 和 Python 环境之间的关系。

简单理解：

| 工具 | 主要解决什么问题 |
| --- | --- |
| `pip` | 安装 Python 包 |
| `venv` | 隔离不同项目的包 |
| `Conda` | 管理 Python 版本和复杂依赖 |
| `uv` | 更快地管理包、环境和项目 |
| `pdb` / VSCode 调试器 | 观察程序运行过程 |
| Jupyter Notebook | 组织讲义、代码和运行结果 |

## 核心问题

Python 环境管理经常围绕四个问题展开：

```text
哪个 python 在运行？
包安装到了哪个环境？
import 时 Python 去哪里找模块？
Notebook 当前使用的是哪个 kernel？
```

后续小节会反复围绕这三个问题展开。
