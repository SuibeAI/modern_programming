# pip 依赖安装

`pip` 是 Python 常用的包安装工具。安装包前，先确认 `pip` 属于哪个 Python 环境。

## 推荐写法

常见写法是直接运行：

```bash
pip install requests
```

但在课程中更推荐使用：

```bash
python3 -m pip install requests
```

这样可以明确表示：用当前 `python3` 对应的 `pip` 来安装包，减少“包安装到了另一个 Python 环境”的问题。

`-m module` 表示把某个 Python 模块当作脚本运行。`pip` 本身也是一个 Python 包，所以 `python3 -m pip` 的意思是：用当前 `python3` 运行它环境中的 `pip` 包。

## 查看已安装包

```bash
python3 -m pip list
python3 -m pip show requests
```

`pip show` 可以查看包版本、安装位置和依赖信息。

## requirements.txt

项目常用 `requirements.txt` 记录依赖：

```text
requests
numpy
```

安装文件中的依赖：

```bash
python3 -m pip install -r requirements.txt
```
