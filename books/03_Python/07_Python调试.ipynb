# Python 调试

调试是定位程序问题的过程。Python 常见调试方式包括命令行中的 `pdb`，以及 VSCode 提供的图形化调试器。

## 什么时候需要调试

当程序没有语法错误，但结果不符合预期时，就需要调试。

常见场景：

- 变量值和自己想的不一样；
- 循环次数不对；
- 分支判断没有进入预期路径；
- 函数返回结果异常。

## pdb：命令行调试

`pdb` 是 Python 内置调试器。可以在希望暂停的位置加入：

```python
breakpoint()
```

Python 3.7 及以上版本中，`breakpoint()` 是内置函数，不需要 `import`。

旧写法是：

```python
import pdb
pdb.set_trace()
```

运行程序时，Python 会在断点处暂停，并进入 `(Pdb)` 交互界面。

## 使用 python -m pdb

如果不想修改代码，可以直接用 `pdb` 启动脚本：

```bash
python -m pdb script.py
```

`-m pdb` 表示把 Python 内置的 `pdb` 模块当作脚本运行，并用它调试后面的 `script.py`。

## pdb 常用命令

| 命令 | 作用 |
| --- | --- |
| `n` | 执行下一行，不进入函数内部 |
| `s` | 执行下一步，进入函数内部 |
| `c` | 继续运行，直到下一个断点或程序结束 |
| `p 变量名` | 打印变量值 |
| `l` | 查看当前位置附近代码 |
| `q` | 退出调试 |

## VSCode 图形化调试

VSCode 调试 Python 代码需要安装 Microsoft 的 Python 插件。安装 Python 插件时，通常会同时安装配套的 Python Debugger。

安装后可以使用图形化断点：

1. 打开 `.py` 文件；
2. 在行号左侧点击，设置红色断点；
3. 点击左侧 Run and Debug 图标；
4. 选择 Python Debugger；
5. 启动调试，程序会在断点处暂停。

暂停后，可以在调试面板中查看变量、调用栈，也可以逐行执行代码。

## 最小例子

创建 `debug_demo.py`：

```python
def add(a, b):
    result = a + b
    return result

x = add(2, 3)
print(x)
```

可以用两种方式调试：

- 在代码中加入 `breakpoint()`，然后运行 `python debug_demo.py`；
- 在 VSCode 中点击行号左侧设置断点，然后启动 Python Debugger。

## 使用建议

`pdb` 适合在终端中快速定位问题。VSCode 图形化调试器更适合观察变量、调用栈和多文件调用关系。

定位问题后，记得删除不需要的 `breakpoint()`。
