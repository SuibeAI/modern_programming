本章介绍 VSCode 在课程中的基本用法：安装 VSCode，打开项目文件夹，理解配置和插件，并使用 AI 编程插件与 SSH 远程开发。

参考材料：[VSCode 官方文档](https://code.visualstudio.com/docs)。

## 本章学习目标

完成本章后，你应该能够：

- 安装 VSCode，并确认 `code` 命令可用。
- 以项目文件夹为单位编辑和运行代码。
- 理解 `settings.json`、用户配置和工作区配置。
- 安装课程相关插件。
- 使用 Copilot 和 Claude Code 辅助编程。
- 使用 Remote - SSH 连接远程服务器。

## 为什么使用 VSCode

VSCode（Visual Studio Code）是微软开发的免费代码编辑器。它适合本课程，主要因为：

| 特性 | 说明 |
| --- | --- |
| 跨平台 | Windows、macOS、Ubuntu 都可使用 |
| 项目导向 | 适合按文件夹组织代码和配置 |
| 内置终端 | 可以在编辑器中直接运行命令 |
| 插件生态 | 支持 Python、Jupyter、Remote - SSH、AI 编程插件 |
| 远程开发 | 可以通过 SSH 操作服务器上的项目 |

## 学习重点

本章不追求覆盖 VSCode 的全部功能，只要求掌握课程中反复用到的工作流：

- 打开项目文件夹；
- 在终端中运行命令；
- 通过 `settings.json` 修改配置；
- 安装必要插件；
- 使用 AI 编程工具；
- 连接远程服务器。
