完成本章后，你应该能够：

- 在自己的电脑上安装并运行 VSCode。

## 为什么选择 VSCode？

VSCode（Visual Studio Code）是微软开发的免费开源代码编辑器。与传统的图形界面编辑器相比，它有以下优势：

| 特性 | 说明 |
| --- | --- |
| **免费开源** | 无需付费，源代码公开 |
| **跨平台** | Windows、macOS、Linux 全支持 |
| **轻量快速** | 启动迅速，资源占用适中 |
| **插件生态** | 超万款插件，扩展能力强 |
| **内置终端** | 无需切换窗口即可运行命令 |
| **Git 集成** | 可视化版本控制操作 |
| **AI 辅助** | 原生支持 Copilot 等编程助手 |
| **远程开发** | 通过 SSH 连接服务器，如本地操作 |

### 基于 VSCode 的衍生产品

VSCode 的成功催生了许多优秀的衍生产品，如 **Cursor**（AI 原生编程环境）、**Trae**（AI 辅助开发工具）等。这些产品都基于 VSCode 的核心架构和插件生态，因此掌握 VSCode 的操作和配置方法，能够让你快速适应这些新兴工具，提升学习迁移价值。

总的来说，VSCode 在易用性、扩展性和性能之间取得了较好的平衡，适合现代编程教学和工程实践。
