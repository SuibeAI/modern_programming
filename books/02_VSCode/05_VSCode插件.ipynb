# VSCode 插件

预计阅读时间：10 分钟。

## 插件是什么

VSCode 基于 Electron 构建，可以简单理解为一个运行在桌面端的网页应用。插件（Extension）是给 VSCode 增加功能的软件包。

插件通过 VSCode 提供的接口注册功能：例如注册一个命令、监听文件保存事件、识别某种语言文件，或在编辑器中提供代码补全。

安装插件后，VSCode 可以：

- 增加新的命令；
- 增加新的设置项；
- 提供代码补全、格式化、调试等能力；
- 和本机工具交互，例如 Python、Git、SSH。

## 如何安装插件

打开插件市场：

| 系统 | 快捷键 |
| --- | --- |
| Windows / Linux | `Ctrl+Shift+X` |
| macOS | `Cmd+Shift+X` |

也可以点击 VSCode 左侧活动栏中的 Extensions 图标。

安装步骤：

1. 在搜索框输入插件名称；
2. 选择发布者可靠、下载量较高的插件；
3. 点击 Install；
4. 如有提示，按要求 Reload 或重启 VSCode。

## 推荐安装

| 插件 | 发布者 | 用途 |
| --- | --- | --- |
| Python | Microsoft | Python 代码运行、调试和环境选择 |
| Pylance | Microsoft | Python 类型分析和智能提示 |
| Jupyter | Microsoft | 打开和运行 `.ipynb` 文件 |
| Remote - SSH | Microsoft | 连接远程服务器开发 |
| WSL | Microsoft | Windows 下连接 WSL 环境 |
| GitHub Copilot | GitHub | AI 代码补全和编程辅助 |
| GitLens | GitKraken | 查看 Git 提交历史和代码作者，可选安装 |

Windows 用户如果使用 WSL，建议安装 `WSL` 插件；macOS 和 Ubuntu 用户通常不需要。

## 检查安装结果

安装后检查：

- 打开 `.py` 文件，确认右下角能选择 Python 解释器；
- 打开 `.ipynb` 文件，确认 VSCode 能显示 notebook；
- 打开命令面板（`Ctrl+Shift+P`），搜索插件命令。插件命令通常写作“插件名: 命令名”，例如 `Python: Select Interpreter`；

## 小结

先安装 Python、Pylance、Jupyter，再根据后续章节需要安装 Remote - SSH、WSL、GitHub Copilot 等插件。
