# VSCode 配置

## VSCode 配置文件

VSCode 的配置并不只保存在一个地方。常见配置主要分为两类：

| 类型 | 位置 | 作用范围 |
| --- | --- | --- |
| 用户配置 | User `settings.json` | 当前用户的所有项目 |
| 工作区配置 | 项目下的 `.vscode/settings.json` | 只影响当前项目 |

用户 `settings.json` 常见位置：

| 系统 | 位置 |
| --- | --- |
| Windows | `%APPDATA%\\Code\\User\\settings.json` |
| macOS | `~/Library/Application Support/Code/User/settings.json` |
| Ubuntu/Linux | `~/.config/Code/User/settings.json` |

工作区配置优先级更高。也就是说，同一个配置项如果在两个地方都设置了，当前项目会优先使用 `.vscode/settings.json`。

## 如何打开 settings.json

打开命令面板：

```text
Ctrl+Shift+P
```

然后搜索：

```text
Preferences: Open User Settings (JSON)
```

如果要修改当前项目配置，搜索：

```text
Preferences: Open Workspace Settings (JSON)
```

## 如何新增配置

`settings.json` 是 JSON 文件。新增配置就是增加一组键值对。例如，将 Tab 设置为 4 个空格，并在保存时自动格式化：

```json
{
  "editor.tabSize": 4,
  "editor.formatOnSave": true
}
```

注意：

- 字符串要用双引号；
- 布尔值写 `true` 或 `false`；
- 最后一项后面不要加逗号。

## 配置有哪些类型

常见配置可以分为：

| 类型 | 示例 | 说明 |
| --- | --- | --- |
| 编辑器配置 | `editor.tabSize` | 控制编辑器行为 |
| 文件配置 | `files.autoSave` | 控制文件保存、编码等 |
| 语言配置 | `[python]` | 只对某种语言生效 |
| 插件配置 | `python.defaultInterpreterPath` | 由插件提供的参数 |
| 远程配置 | Remote SSH 相关配置 | 影响远程开发 |

插件安装后，通常会新增自己的配置项。例如 Python 插件会增加解释器、测试、格式化等相关参数。

## 语言级配置

有些设置只希望对某种语言生效，可以写成：

```json
{
  "[python]": {
    "editor.formatOnSave": true,
    "editor.tabSize": 4
  }
}
```

这表示这些设置只影响 Python 文件。

## 代理和环境变量

`http_proxy`、`https_proxy` 等环境变量会影响很多联网行为，例如下载插件、Git 访问网络、终端里运行 `pip` 或 `curl`。

但要注意：VSCode 图形界面不一定继承你终端里的环境变量。也就是说，在终端里设置了代理，不一定代表 VSCode 插件市场也能使用这个代理。

VSCode 自身也有代理配置，例如配置VSCode使用本机代理：

```json
{
  "http.proxy": "http://127.0.0.1:7890",
}
```

> 注意：只有在网络确实受限时才需要配置代理。不要随意复制不理解的代理地址。

## 需要备份吗？

用户配置建议开启 VSCode 的 Settings Sync，同步到自己的账号。

项目配置 `.vscode/settings.json` 是否提交到 Git，要看内容：

- 团队共享规则可以提交，例如格式化、缩进、推荐解释器路径；
- 个人路径、代理地址、密钥不要提交；
- 不确定时，先不要提交，或问团队成员。

## 小结

VSCode 配置的核心是 `settings.json`。先区分用户配置和工作区配置，再判断配置属于编辑器、语言、插件还是网络环境。理解这个逻辑，比记住某个具体设置更重要。

更多内容可参考 [VSCode 官方设置文档](https://code.visualstudio.com/docs/configure/settings)。
