# GitHub Copilot 编程插件

预计阅读时间：10 分钟。

GitHub Copilot 是 VSCode 中常用的 AI 编程插件，可以提供代码补全、代码解释、修改建议和对话式编程辅助。

## 安装

1. 打开插件市场：`Ctrl+Shift+X`，macOS 使用 `Cmd+Shift+X`；
2. 搜索 `GitHub Copilot`；
3. 确认发布者为 GitHub；
4. 点击 Install；
5. 按提示登录 GitHub 并完成授权。

安装完成后，VSCode 状态栏通常会出现 Copilot 图标。

## 代码补全

在 Python 文件中输入函数名或注释，Copilot 会根据上下文给出灰色补全建议。

```python
# 判断一个数字是否为素数
def is_prime(n):
```

常用操作：

| 操作 | 说明 |
| --- | --- |
| `Tab` | 接受当前建议 |
| `Esc` | 取消当前建议 |
| 继续输入 | 用更多上下文引导 Copilot |

## 代码 Agent

Copilot 不只会补全代码，也可以作为代码 Agent 使用。常见方式是打开 Copilot Chat，然后用自然语言描述任务。

简单任务可以这样提问：

- 解释当前文件的主要功能；
- 为这个函数补充简单的测试用例；
- 检查这段代码是否有明显错误。

也可以给它一个更完整的编程任务，例如：

```text
用 PyTorch 写一个简单的自注意力机制类 SelfAttention，输入形状为 [batch, seq_len, dim]，输出形状保持不变。
请同时写一段测试代码，随机生成输入张量，并打印输入和输出的 shape。
```

使用 Agent 结果时，要先阅读再运行，不要直接接受看不懂的修改。

## 功能例子

| 场景 | 示例 |
| --- | --- |
| 生成函数 | 根据注释生成 `is_prime`、`gcd` 等小函数 |
| 解释代码 | 选中代码后让 Copilot 解释逻辑 |
| 修改代码 | 要求 Copilot 将变量名改得更清楚 |
| 生成测试 | 为已有函数生成简单 `assert` 测试 |
| 排查错误 | 把报错信息粘贴给 Copilot，请它分析原因 |

课堂练习中，优先让 Copilot 处理小范围、可验证的任务。

## 小结

Copilot 适合用于代码补全、解释、测试生成和小范围修改。它能提高效率，但生成结果仍需要人工检查和运行验证。
