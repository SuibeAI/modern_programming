# VSCode SSH 远程开发

预计阅读时间：10 分钟。

本节只介绍两件事：为什么使用 SSH，以及如何用 VSCode 连接远程服务器。

## 为什么使用 SSH

SSH（Secure Shell）可以让本地电脑安全连接到远程服务器。对开发来说，常见原因包括：

- 远程服务器有更强的 CPU、GPU、内存或存储；
- 代码和数据已经在服务器上，不需要频繁下载；
- 服务器环境更接近真实运行环境；
- VSCode 可以像打开本地文件夹一样打开远程文件夹。

连接后，编辑器在本地显示，代码、终端和运行环境在远程服务器上。

## 工作原理

Remote - SSH 的基本过程是：

1. 本地 VSCode 通过 SSH 登录远程服务器；
2. VSCode 在远程服务器上启动一个 VSCode Server；
3. 本地 VSCode 负责显示界面，远程 VSCode Server 负责读写文件和运行命令。

因此，你看到的是本地 VSCode 界面，但实际操作的是远程服务器上的项目。

## 安装 Remote - SSH 插件

打开 VSCode 插件市场，搜索并安装：

```text
Remote - SSH
```

发布者选择 Microsoft。安装完成后，VSCode 左侧通常会出现远程开发图标。

## 配置 SSH 主机

打开命令面板：

```text
Ctrl+Shift+P
```

搜索并选择：

```text
Remote-SSH: Open SSH Configuration File
```

选择 `~/.ssh/config`，添加服务器配置：

```text
Host myserver
    HostName 服务器地址
    User 用户名
    Port 22
    IdentityFile ~/.ssh/id_rsa
```

其中 `Host myserver` 是本地别名，可以自行命名；`HostName`、`User`、`Port` 和 `IdentityFile` 根据服务器信息填写。

## 使用 VSCode 连接

打开命令面板，搜索：

```text
Remote-SSH: Connect to Host
```

选择刚才配置的 `myserver`。

首次连接时，VSCode 会在远程服务器上安装 VSCode Server。连接成功后，左下角会显示类似：

```text
SSH: myserver
```

然后点击 Open Folder，选择远程服务器上的项目目录。此时 VSCode 中看到的是远程文件，终端也运行在远程服务器上。

> 注意：连接远程服务器后，有些插件需要安装到远程端才会生效。例如 Python、Jupyter 这类和运行环境相关的插件，通常需要在 `SSH: myserver` 环境中重新安装。

## 小结

VSCode Remote - SSH 的核心流程是：安装插件，配置 `~/.ssh/config`，连接主机，打开远程项目文件夹。

完成后，你可以在本地 VSCode 中编辑远程服务器文件，并在远程终端中运行命令。
