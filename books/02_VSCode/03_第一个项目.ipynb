# 第一个 VSCode 项目

预计阅读与练习时间：10 分钟。

本节完成一个最小项目：打开文件夹，新建 Python 文件，并在 VSCode 终端中运行。

## VSCode 以项目为导向

VSCode 通常以文件夹作为项目。打开文件夹后，左侧资源管理器显示项目文件，内置终端默认也位于这个项目目录。

## 准备项目文件夹

在系统终端中创建练习目录，并用 VSCode 打开：

```bash
mkdir ~/vscode-first-project
cd ~/vscode-first-project
code .
```

`code .` 表示用 VSCode 打开当前文件夹。

## 新建 Python 文件

在左侧资源管理器中新建文件：

```text
hello.py
```

写入并保存：

```python
print('hi')
```

## 在终端中运行

在 VSCode 中打开终端：Terminal -> New Terminal，或使用快捷键 `` Ctrl+` ``。

先确认当前目录下能看到 `hello.py`：

```bash
ls
```

然后运行：

```bash
python3 hello.py
```

如果运行成功，会看到：

```text
hi
```

如果没有 `python3` 命令，可以尝试：

```bash
python hello.py
```

## 小结

VSCode 的基本工作方式是：打开项目文件夹，在文件夹中编辑文件，并在内置终端中运行命令。

更多内容可参考 [VSCode 官方入门文档](https://code.visualstudio.com/docs/editing/getting-started)。
