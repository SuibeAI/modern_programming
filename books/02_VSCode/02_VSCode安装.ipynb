# VSCode 安装与验证

预计阅读与练习时间：10-20 分钟。

本节介绍如何安装 Visual Studio Code（简称 VSCode），并确认命令行中的 `code` 命令可用。

参考资料：[VSCode 官方安装文档](https://code.visualstudio.com/docs/setup/setup-overview)。

## 学习目标

- 在自己的电脑上安装 VSCode。
- 理解 Windows、macOS、Ubuntu 三种系统的安装差异。
- 使用 `code --version` 验证安装结果。
- 使用 `code .` 从当前目录打开 VSCode。

## Windows：安装 VSCode

Windows 上推荐使用官网安装包。这个方式最直接，也便于确认 `code` 命令是否加入 `PATH`。

1. 打开 [VSCode 官网](https://code.visualstudio.com/)。
2. 下载 Windows 安装包。一般选择 **User Installer**，不需要管理员权限。
3. 双击安装包，按向导完成安装。
4. 安装选项中建议勾选：
   - `Add to PATH`：允许在 PowerShell、Windows Terminal 中使用 `code` 命令。
   - `Register Code as an editor for supported file types`：把 VSCode 注册为常见代码文件的编辑器。
   - `Add "Open with Code" action`：在右键菜单中添加“用 Code 打开”。
5. 安装完成后，重新打开终端，再验证 `code` 命令。

## macOS：安装 VSCode

macOS 上推荐使用官网安装包。安装完成后，如果需要在终端中使用 `code` 命令，需要在 VSCode 中额外启用一次。

1. 打开 [VSCode 官网](https://code.visualstudio.com/)。
2. 下载 macOS 安装包。
3. 解压或打开安装包。
4. 将 Visual Studio Code 拖入 `Applications` 文件夹。
5. 从 Launchpad 或 `Applications` 中打开 VSCode。

首次打开时，macOS 可能会提示安全确认，选择“打开”即可。

如果终端中无法运行 `code`，在 VSCode 中执行：

1. 按 `Cmd+Shift+P` 打开命令面板。
2. 输入 `Shell Command`。
3. 选择 `Shell Command: Install 'code' command in PATH`。
4. 重新打开终端，再运行 `code --version`。

## Ubuntu：安装 VSCode

Ubuntu 桌面系统推荐从官网下载安装包。VSCode 官网会提供 `.deb` 文件，适合 Ubuntu 和 Debian 系统。

1. 打开 [VSCode 官网](https://code.visualstudio.com/)。
2. 下载 `.deb` 安装包。
3. 在文件管理器中双击 `.deb` 文件安装。
4. 如果图形界面安装失败，可以在下载目录运行：

```bash
sudo apt install ./code_*.deb
```

这条命令会修改系统软件安装状态，需要管理员权限。执行前确认当前目录中只有你刚下载的 VSCode `.deb` 安装包。

## 验证安装

安装完成后，重新打开终端（PowerShell、Windows Terminal、Terminal 或 Bash），运行：

```bash
code --version
```

### 应该看到什么

如果安装成功，会看到类似输出：

```
1.xx.x
xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx
x64
```

第一行是版本号，后面是提交哈希和系统架构。不同机器上的输出不同，这是正常现象。

### 如果提示 "command not found"

- Windows：重新打开终端；若仍失败，重新安装并确认勾选 `Add to PATH`。
- macOS：按上文安装 `code` 命令后，重新打开终端。
- Ubuntu：确认 VSCode 已安装，并检查 `which code` 是否有输出。

### 测试打开 VSCode

在终端中运行：

```bash
code .
```

这会打开 VSCode，并显示当前目录的内容。

## 常见问题

| 现象 | 可能原因 | 处理方式 |
| --- | --- | --- |
| `code: command not found` | `code` 命令没有加入 `PATH` | 按本节对应系统的说明重新配置 |
| Windows 中 `code .` 没反应 | 终端还没读取新的环境变量 | 关闭并重新打开终端 |
| 下载速度慢 | 网络或镜像问题 | 记录报错，课堂统一处理 |

本节不展开复杂镜像源配置。安装软件前要确认下载来源可靠，课堂中优先使用 VSCode 官网。
