# Claude Code 编程插件

预计阅读时间：10 分钟。

Claude Code 是 VSCode 中常用的 AI 编程工具。它可以作为代码 Agent 理解项目文件、修改代码，并根据需要调用终端命令。

## 和 Copilot 的区别

| 工具 | 主要方式 | 适合场景 |
| --- | --- | --- |
| GitHub Copilot | 编辑器内代码补全和聊天 | 补全函数、解释小段代码 |
| Claude Code | VSCode 插件 + 终端 Agent | 理解项目、修改多个文件、运行命令 |

简单理解：Copilot 更偏向代码补全，Claude Code 更偏向在项目中完成多步骤修改任务。

## 安装 Claude Code

先在 VSCode 插件市场中搜索 `Claude Code` 并安装插件。

Claude Code 还需要本机可以使用 `claude` 命令。如果当前环境还没有安装，可以在 VSCode 终端中运行：

```bash
npm install -g @anthropic-ai/claude-code
```

安装后检查：

```bash
claude --version
```

如果 `npm` 不可用，需要先安装 Node.js。

## 配置 GLM 后端

课堂中如果使用 GLM 作为 Claude Code 后端，需要配置 Claude Code 访问智谱提供的 Anthropic-compatible 接口。

编辑或新建 Claude Code 配置文件：

```text
~/.claude/settings.json
```

写入以下配置，并把 `your_zhipu_api_key` 替换为自己的 API Key：

```json
{
  "env": {
    "ANTHROPIC_AUTH_TOKEN": "your_zhipu_api_key",
    "ANTHROPIC_BASE_URL": "https://open.bigmodel.cn/api/anthropic",
    "API_TIMEOUT_MS": "3000000",
    "CLAUDE_CODE_DISABLE_NONESSENTIAL_TRAFFIC": "1"
  }
}
```

再编辑或新建：

```text
~/.claude.json
```

写入：

```json
{
  "hasCompletedOnboarding": true
}
```

配置完成后，重新打开一个新的 VSCode 终端，让配置生效。

## 在 VSCode 中启动

先用 VSCode 打开项目文件夹，然后使用 `` Ctrl+Shift+` `` 打开内置终端。

确认终端位于项目目录：

```bash
pwd
ls
```

启动 Claude Code：

```bash
claude
```

如果提示是否信任当前文件夹，确认后再继续。

## 小结

Claude Code 更适合在项目目录中完成多步骤任务。使用顺序是：先安装 Claude Code，再配置模型后端，最后进入项目目录启动并完成可检查、可运行的小任务。

更多内容可参考 [BigModel Claude Code 文档](https://docs.bigmodel.cn/cn/guide/develop/claude)。
