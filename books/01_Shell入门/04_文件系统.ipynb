# Shell 中的文件系统

命令行工作的第一条主线是文件系统。你需要知道自己在哪里、文件在哪里，以及命令会改变哪些文件。

## 学习目标

完成本节后，你应该能够：

- 使用 `pwd` 确认当前位置。
- 使用 `cd` 在目录之间移动。
- 区分绝对路径、相对路径、`.`、`..` 和 `~`。
- 使用 `ls`、`mkdir`、`touch`、`cp`、`mv`、`rm` 完成基础文件操作。
- 在执行文件操作后用 `pwd` 和 `ls -la` 检查结果。

## 核心概念

- 当前工作目录：命令默认发生的位置，可用 `pwd` 查看。
- 绝对路径：从根目录 `/` 开始，例如 `/home/username/course` 或 `/tmp/demo`。
- 相对路径：从当前目录出发，例如 `../data` 或 `./scripts/run.sh`。
- `.`：当前目录。
- `..`：上一级目录。
- `~`：当前用户的家目录。
- 隐藏文件：以 `.` 开头的文件或目录，例如 `.gitignore`、`.ssh`。

## 查看和切换位置

`pwd` 是 print working directory 的缩写，用来显示当前工作目录。

```bash
pwd
```

`cd` 是 change directory 的缩写，用来切换目录。

```bash
cd ..
cd ~
cd ./books
```

## 查看目录内容

`ls` 用来列出目录内容。常用 `-l` 查看详细信息，`-a` 显示隐藏文件。

```bash
ls
ls -la
ls -la ./
```

## 目录操作

`mkdir` 用来创建目录。下面的命令会创建一个名为 `demo` 的目录。

```bash
mkdir demo
```

`rmdir` 用来删除空目录。它只能删除空目录，因此比直接递归删除更适合初学者练习。

```bash
rmdir demo
```

## 文件操作

常见文件操作包括创建、复制、移动、重命名和删除。下面每个命令都建议执行后用 `ls -la` 检查结果。

`touch` 可以创建一个空文件。如果文件已存在，它会更新文件的修改时间。

```bash
touch example.txt
```

`cp` 用来复制文件。下面的命令会把 `example.txt` 复制成 `copy.txt`。

```bash
cp example.txt copy.txt
```

`mv` 可以移动文件，也可以重命名文件。下面的命令会把 `copy.txt` 重命名为 `backup.txt`。

```bash
mv copy.txt backup.txt
```

`rm` 会删除文件。删除前先用 `ls -la` 确认文件名，避免误删。

```bash
rm example.txt
rm backup.txt
```

## 操作后的检查

每做一次文件操作，都建议用下面两条命令检查：

```bash
pwd
ls -la
```

`pwd` 确认你在哪里，`ls -la` 确认文件是否按预期变化。后续学习 Git 时，很多问题本质上也是“我在哪个目录，哪些文件变了”。

## 小练习

在一个临时目录里完成：

```bash
mkdir shell-demo
cd shell-demo
pwd
touch notes.txt
cp notes.txt notes-copy.txt
mv notes-copy.txt backup.txt
ls -la
```

练习结束后，说明每条命令改变了什么。确认无误后，可以回到上一级目录再删除临时目录。`rm -r` 会递归删除目录，执行前必须确认目录名是 `shell-demo`。

```bash
cd ..
rm -r shell-demo
```

## 小结

文件系统操作的关键是先确认位置，再执行操作，最后检查结果。记住三个问题：我在哪里？我要操作哪个文件或目录？命令执行后发生了什么变化？
