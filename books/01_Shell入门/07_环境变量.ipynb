# 环境变量

环境变量是 Shell 中的键值配置，会影响子进程的行为。与命令参数不同，被 `export` 的环境变量会传递给之后启动的程序。本节按用途分类介绍常用的环境变量。

## 查看环境变量

### 查看单个变量

用 `echo $变量名` 查看某个变量的值：

```bash
echo $PATH
echo $HOME
echo $SHELL
```

### 查看所有环境变量

用 `env` 命令查看当前所有环境变量（按行列出）：

```bash
env
env | grep PYTHON  # 只显示包含 PYTHON 的变量
```

这在诊断环境问题时很有用。例如，若某个程序找不到，可以先检查 `env` 是否包含所需的环境变量。

## 路径相关

### `PATH`

当你输入 `python3`、`git` 等命令时，Shell 会沿着 `PATH` 中列出的目录查找同名可执行文件。如果程序已安装但命令找不到，通常是 `PATH` 配置不对。

```bash
echo $PATH
which git
which python3
```

`PATH` 由多个目录组成，用 `:` 分隔。例如 `/usr/local/bin:/usr/bin:/bin`。

### `PYTHONPATH`

Python 运行时会沿着 `PYTHONPATH` 搜索模块。当你想使用自己编写的模块或非标准库的包时，可以用它添加搜索路径。

```bash
export PYTHONPATH=/path/to/your/module:$PYTHONPATH
python3 -c "import sys; print(sys.path)"
```

这样 Python 会把当前 `PYTHONPATH` 里的目录添加到模块搜索路径的开头。

### `LD_LIBRARY_PATH`

Linux 程序运行时需要加载共享库（.so 文件）。`LD_LIBRARY_PATH` 指定库的搜索目录。使用 CUDA、TensorRT 等库时经常需要设置。

```bash
export LD_LIBRARY_PATH=/usr/local/cuda/lib64:$LD_LIBRARY_PATH
```

## 网络代理

如果网络需要通过代理服务器访问，可以设置以下变量。大多数支持网络的命令（curl、wget、pip 等）都会读取这些变量。

```bash
export http_proxy=http://proxy.example.com:8080
export https_proxy=http://proxy.example.com:8080
export no_proxy=localhost,127.0.0.1
```

`no_proxy` 用来指定哪些地址不走代理。多个地址用 `,` 分隔。

## 深度学习相关

### 镜像与源

使用国内镜像可以加快下载速度。这些环境变量被 pip、conda、huggingface-hub 等工具读取。

```bash
# PyPI 镜像（清华源）
export PIP_INDEX_URL=https://pypi.tsinghua.edu.cn/simple

# Hugging Face 模型库镜像
export HF_ENDPOINT=https://hf-mirror.com

# Conda 频道（在 ~/.condarc 中配置更常见，但也可用环境变量）
export CONDA_CHANNELS=https://mirrors.tsinghua.edu.cn/anaconda/cloud
```

### `CUDA_VISIBLE_DEVICES`

指定 GPU 设备。在多 GPU 机器上很常用：

```bash
# 只使用第 0 和 2 号 GPU
export CUDA_VISIBLE_DEVICES=0,2

# 只使用第 1 号 GPU
export CUDA_VISIBLE_DEVICES=1
```

不设置此变量时，程序会使用所有可见的 GPU。

### 日志与调试

```bash
# TensorFlow：设置日志级别（0=DEBUG, 1=INFO, 2=WARNING, 3=ERROR）
export TF_CPP_MIN_LOG_LEVEL=2

# PyTorch：让 CUDA 操作同步执行，便于定位 GPU 报错位置
export CUDA_LAUNCH_BLOCKING=1

# Python：不生成 .pyc 文件
export PYTHONDONTWRITEBYTECODE=1
```

## 临时设置与永久设置

### 临时设置

在当前 Shell 会话中设置变量（只对当前终端窗口有效）：

```bash
export MY_VAR=value
echo $MY_VAR
```

关闭终端后，这个设置就消失了。

### 永久设置

将设置写入 Shell 配置文件，这样每次打开新终端都会加载。

**第一步：确定当前 Shell 类型**

```bash
echo $SHELL
```

输出可能是 `/bin/bash`、`/bin/zsh`、`/bin/sh` 等。

**第二步：根据 Shell 类型选择配置文件**

| Shell 类型 | 配置文件 |
| --- | --- |
| bash | `~/.bashrc` 或 `~/.bash_profile` |
| zsh | `~/.zshrc` |
| sh | `~/.profile` |

大多数现代 Linux 发行版和 macOS 默认用 bash 或 zsh。

**第三步：添加配置**

根据上一步确定的 Shell 类型，选择对应的配置文件并添加环境变量。有两种方法：

**方法 A：用命令追加（快速）**

```bash
# 如果用 bash
echo 'export MY_VAR=value' >> ~/.bashrc
source ~/.bashrc

# 如果用 zsh
echo 'export MY_VAR=value' >> ~/.zshrc
source ~/.zshrc
```

**方法 B：用编辑器手动添加（推荐，更清楚）**

```bash
# 用编辑器编辑（例如 vi）
vi ~/.bashrc  # 或用其他编辑器如 nano
```

直接编辑的好处是可以看到整个配置文件，避免重复添加相同的 `export` 语句。

用 `export` 关键字设置环境变量，这样它会传递给所有子进程。不用 `export` 只是设置当前 Shell 的变量，子进程看不到。

### 小结

- 环境变量影响所有子进程的行为。
- `PATH` 和 `PYTHONPATH` 用来指定搜索路径；`export` 使变量对子进程可见。
- 网络代理、镜像、GPU 配置等都通过环境变量来控制。
- 临时设置用 `export`；永久设置写入 Shell 配置文件。
