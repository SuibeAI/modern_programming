# Shell 与命令结构

预计阅读和练习时间：15-20 分钟。

本节解释 Shell、终端和命令之间的关系，并建立阅读一条命令的基本方法。学完后，你应该能说清楚：一条命令由哪些部分组成，Shell 如何找到要运行的程序，以及遇到陌生命令时如何查帮助。

## 学习目标

完成本节后，你应该能够：

- 区分终端（Terminal）、Shell 和命令行程序。
- 拆解一条命令中的命令名、选项和参数。
- 使用 `which`、`type` 判断命令来自哪里。
- 理解 `PATH` 为什么会影响命令查找。
- 使用 `--help`、`man` 和 `tldr` 查询命令用法。
- 知道在 Ubuntu 中如何安装新的命令行工具。

## 什么是 Shell

Shell 是用户与操作系统之间的一层命令行接口。用户在终端中输入命令后，Shell 负责解析命令文本，处理变量替换、通配符展开、输入输出重定向等操作，找到对应的可执行程序，传递参数并启动程序，最后把程序的输出结果显示在终端中。

```{mermaid}
flowchart LR
    A[用户输入命令] --> B[终端 Terminal]
    B --> C[Shell 解析命令]
    C --> D[查找命令程序]
    D --> E[启动进程]
    E --> F[读取/修改系统资源]
    E --> G[输出结果]
    G --> B
```

三者的分工可以这样理解：

| 名称 | 负责什么 | 例子 |
| --- | --- | --- |
| 终端 | 显示文本、接收键盘输入 | Windows Terminal、Ubuntu Terminal、macOS Terminal |
| Shell | 解释命令文本 | `bash`、`zsh` |
| 命令行程序 | 真正完成任务 | `ls`、`git`、`python3`、`curl` |

当你输入 `ls -l` 时：

- 终端只是把这行文本交给 Shell；
- Shell 找到 `ls` 程序并启动它；
- `ls` 程序读取当前目录的信息，并把文件列表、权限、大小和修改时间等结果输出到终端。

## 为什么需要 Shell

命令行界面和图形化界面，是我们与操作系统交互的两种主要方式。图形化界面更直观，适合探索和一次性操作；命令行更精确，适合记录、复用和自动化。

为什么有了图形化界面之后，还需要命令行呢？

1. 命令可以被保存、复用和组合。把一组命令写成脚本后，同一个流程可以反复执行，结果也更容易检查。图形化界面依赖人的点击和选择，步骤多时更容易遗漏或出错。
2. 生产环境通常没有图形界面。很多应用运行在远程服务器、容器或云主机上，我们主要通过 Shell 登录、部署、查看日志和排查问题。

此外，基于大模型的 AI 智能体如果具备执行 Shell 命令的能力，就可以读写文件、运行程序、安装依赖、执行测试，进而在一定范围内操控计算机。理解 Shell，有助于我们判断 AI 智能体能做什么、不能做什么，以及哪些操作需要人工确认。

## 一条命令的结构

一条命令通常由三部分组成：

```bash
命令名 [选项] [参数]
```

例如：

```bash
ls -l ./
```

| 部分 | 示例 | 作用 |
| --- | --- | --- |
| 命令名 | `ls` | 要运行的程序。很多命令名来自英文缩写，例如 `ls` 来自 list |
| 选项 | `-l` | 改变程序行为，这里表示长格式（long format）输出 |
| 参数 | `./` | 程序处理的对象，这里表示当前目录 |



## 命令在哪里

Shell 运行命令前，需要先找到命令对应的程序。可以用 `which` 查看可执行文件路径：

```bash
which python3
which git
```

输出可能类似：

```text
/usr/bin/python3
/usr/bin/git
```

还可以用 `type` 查看 Shell 如何理解某个名字：

```bash
type cd
type ls
type python3
```

`cd` 通常是 Shell 内置命令（shell builtin），而 `python3` 通常是某个目录下的可执行文件。这个区别解释了为什么有些命令（例如 `cd`）没有独立文件路径。

## `PATH` 环境变量

为什么我们只输入 `python3`，Shell 就能找到 `/usr/bin/python3` 这样的程序？原因是 Shell 会在 `PATH` 环境变量列出的目录中依次查找。

查看当前 `PATH`：

```bash
echo $PATH
```

`PATH` 的值通常是一串用冒号分隔的目录，例如：

```text
/usr/local/bin:/usr/bin:/bin
```

实际操作中，`which python3` 可以帮助判断自己运行的是哪一个 Python。尤其是安装了 Conda、系统 Python、课程虚拟环境之后，很多问题都来自“你以为运行的是 A，其实 Shell 找到的是 B”。

本节只要求理解 `PATH` 的作用；环境变量的更多内容会在 [06_环境变量.ipynb](06_环境变量.ipynb) 中展开。

## 查看命令帮助

遇到陌生命令时，优先查它自己的帮助信息。

```bash
ls --help
man ls
tldr ls
```

常用方式：

| 方法 | 适合场景 |
| --- | --- |
| `命令 --help` | 快速查看常用选项 |
| `man 命令` | 查看系统手册，内容更完整 |
| `tldr 命令` | 查看社区整理的简短示例，需要额外安装 |

阅读帮助时，先找三类信息：这个命令做什么、常用选项是什么、示例里如何使用。

## 安装新的命令

Ubuntu 已经内置了很多命令，但有些工具需要额外安装。常用方式是使用 `apt` 包管理器。

先更新软件包索引：

```bash
sudo apt update
```

再安装一个简单工具，例如 `tree`：

```bash
sudo apt install tree
```

安装后验证：

```bash
tree --version
tree ./
```

`sudo` 表示以管理员权限执行命令。输入密码时终端通常不会显示字符，这是正常现象。安装软件前要确认命令来源可靠，课堂中只安装老师明确要求的工具。

如果 `apt update` 或 `apt install` 下载速度太慢，通常需要更换国内镜像源，本节不展开。感兴趣的同学可以自行搜索。

## 小练习

在 Ubuntu 终端中运行：

```bash
pwd
ls -la
type ls
type cd
echo $PATH
ls --help
man ls
```

对每条命令思考：

- 命令名是什么？
- 是否有选项？
- 是否有参数？
- 输出告诉了你什么？

不要急着记住所有选项。先练习把命令拆开读懂，这是后面学习文件系统、管道和脚本的基础。

## 小结
