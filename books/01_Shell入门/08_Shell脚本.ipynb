# Shell 脚本

预计阅读与练习时间：10-20 分钟。

当一组命令需要反复执行，就应该考虑写成 Shell 脚本。脚本让流程可以被复用、检查和改进。

## 学习目标

完成本节后，你应该能够：

- 知道什么时候适合把命令整理成脚本。
- 读懂一个最小 Shell 脚本的基本结构。
- 使用 `chmod +x` 和 `./脚本名` 运行脚本。
- 用脚本保存环境检查和重复实验命令。

## 什么时候写脚本

适合写 Shell 脚本的场景：

- 重复整理文件或目录。
- 批量查看、筛选、转换文本。
- 自动运行一组项目命令。
- 用不同参数重复运行同一个程序。

如果任务需要复杂数据结构、长期维护的业务逻辑或图形界面，通常应考虑 Python、JavaScript 或其他通用语言。

## 一个脚本应该说明什么

脚本通常需要包含：

- 明确的输入：参数、环境变量或固定目录。
- 清晰的输出：生成文件、打印摘要或返回状态码。
- 基本错误处理：命令失败时停止，避免继续产生错误结果。
- 帮助信息：让别人知道如何运行。

## 最小脚本示例

先进入本章目录：

```bash
cd books/01_Shell入门
```

创建 `check_env.sh`，用来检查当前 Shell 环境：

```bash
#!/usr/bin/env bash
set -euo pipefail

echo "Current directory: $(pwd)"
echo "PATH: ${PATH}"
echo "PYTHONPATH: ${PYTHONPATH:-<not set>}"
echo "Git version: $(git --version)"
echo "Python version: $(python3 --version)"
echo "python3 path: $(which python3)"
echo "pip3 path: $(which pip3 || echo '<not found>')"
```

这个脚本里有几个常见写法：

| 写法 | 作用 |
| --- | --- |
| `#!/usr/bin/env bash` | 指定用 Bash 运行脚本 |
| `set -euo pipefail` | 遇到常见错误时尽早停止 |
| `$(命令)` | 先运行命令，再把输出放到当前位置 |
| `${PYTHONPATH:-<not set>}` | 如果 `PYTHONPATH` 没有设置，就显示 `<not set>` |
| `which python3` | 查看 `python3` 实际对应的程序路径 |



运行前需要赋予执行权限：



```bash
chmod +x check_env.sh
./check_env.sh
```



运行后会看到当前目录、`PATH`、`PYTHONPATH`、Git 版本、Python 版本，以及 `python3` 和 `pip3` 对应的程序路径。不同机器上的输出通常不同，这是正常现象。

`set -euo pipefail` 能让脚本在常见错误时尽早停止，适合课程中的自动化练习。

也可以不用执行权限，直接用 `bash` 运行脚本：

```bash
bash ./check_env.sh
```

## 重复跑实验 3 次

假设需要用不同参数运行 `train.py`，并分别保存结果，可以用 Shell 脚本把这些命令固定下来。这样既不容易漏跑，也方便之后复现实验。

例如创建 `run_experiments.sh`：

```bash
#!/usr/bin/env bash
set -euo pipefail

mkdir -p results

python3 train.py --lr 0.1 > results/lr_0.1.txt
python3 train.py --lr 0.01 > results/lr_0.01.txt
python3 train.py --lr 0.001 > results/lr_0.001.txt
```

运行方式：

```bash
chmod +x run_experiments.sh
./run_experiments.sh
```

这里的 `train.py` 是假设存在的训练脚本，`--lr` 表示传给训练脚本的学习率参数。`>` 会把每次实验的标准输出写入对应文件。如果文件已经存在，会覆盖原内容；正式实验前应确认结果文件名。

## 常见问题

| 现象 | 可能原因 | 检查方式 |
| --- | --- | --- |
| `Permission denied` | 脚本没有执行权限 | 运行 `chmod +x 脚本名` |
| `command not found` | 命令没有安装，或不在 `PATH` 中 | 运行 `which 命令名` |
| 输出被覆盖 | 使用了 `>` 写入已有文件 | 写入前先 `ls` 检查文件名 |

- 脚本名为 `check_env.sh`。
- 至少打印当前目录、Git 版本、Python 版本、`PATH`、`PYTHONPATH`、`python3` 路径和 `pip3` 路径。
- 能通过 `./check_env.sh` 运行。

完成后，把脚本内容和运行输出保存到课程笔记中。

### 小结

- Shell 脚本适合保存重复执行的命令流程。
- `#!/usr/bin/env bash` 指定用 Bash 运行脚本。
- `chmod +x 脚本名` 赋予执行权限，`./脚本名` 从当前目录运行脚本。
- 脚本中使用 `>` 写入结果文件时，要注意可能覆盖旧文件。
