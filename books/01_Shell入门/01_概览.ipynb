# Shell 入门概览

本章帮助你建立命令行工作流的基本认识：先准备 Shell 环境，再理解命令如何运行，最后学习文件系统、管道、环境变量和脚本化自动化。

参考材料：《计算机教育中缺失的一课》Shell 主题：[课程讲义](https://missing-semester-cn.github.io/2026/course-shell/)。

## 本章学习目标

完成本章后，你应该能够：

- 在自己的电脑上打开并验证可用的 Shell 环境。
- 读懂一条命令的命令名、选项、参数、输入和输出。
- 使用路径、当前工作目录和常见文件命令管理项目文件。
- 用管道和重定向组合小工具处理文本。
- 解释 `PATH`、`HOME` 等环境变量为什么会影响命令行为。
- 判断什么时候应该把重复命令整理成 Shell 脚本。
