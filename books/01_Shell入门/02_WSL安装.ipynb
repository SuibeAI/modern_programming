# 安装 Shell 环境

预计阅读和操作时间：15-20 分钟。

本节目标是让每位同学都有一个可复现的命令行环境。Windows 用户推荐使用 WSL（Windows Subsystem for Linux），macOS 和 Linux 用户可以使用系统自带终端。

## 学习目标

完成本节后，你应该能够：

- 打开一个可用的 Shell。
- 判断自己正在使用 Windows、macOS、Linux 还是 WSL。
- 验证 `bash`、`git`、`python` 等课程常用命令是否可用。
- 记录环境问题，便于课堂上快速定位。

## 不同系统的建议

| 系统 | 推荐环境 | 说明 |
| --- | --- | --- |
| Windows | WSL + 默认 Ubuntu（当前通常为 Ubuntu 24.04） | 最接近 Linux 服务器和课程示例 |
| macOS | Terminal 或 iTerm2 | 系统自带 zsh，可安装 bash/git/python |
| Linux | 系统终端 | 通常已经具备课程需要的大部分工具 |

Windows 上的 PowerShell 和 CMD 也能完成部分任务，但路径、命令和脚本行为与 Linux Shell 有差异。本课程为了减少环境差异，Windows 用户优先使用 WSL。

## Windows：安装 WSL

Windows 上建议先启用 WSL 需要的系统功能，再安装 Ubuntu。完整流程如下。

### 第一步：启用 Windows 功能

打开“启用或关闭 Windows 功能”，勾选：

- 适用于 Linux 的 Windows 子系统
- 虚拟机平台

![Windows 功能设置](figures/01_01_windows功能设置.png)

勾选完成后，按系统提示重启电脑。

### 第二步：安装 Ubuntu

安装 Ubuntu 有两种常见方式。课堂优先使用方法一。

#### 方法一：通过 Microsoft Store 安装 Ubuntu（推荐）

打开 Microsoft Store，搜索并安装 Ubuntu。当前默认版本通常为 Ubuntu 24.04 LTS。

![Microsoft Store 安装 Ubuntu](figures/01_02_Windows商店安装Ubuntu.png)

如果 Microsoft Store 无法正常下载，可以改用 WSL 官方发布页面下载安装包：

- 官方地址：[https://github.com/microsoft/WSL/releases](https://github.com/microsoft/WSL/releases)
- 课程示例安装包：`wsl.2.7.8.0.x64.msi`

![下载wsl离线安装包](figures/01_04_wsl_download.png)
下载后双击 `.msi` 安装包，按提示完成安装。

#### 方法二：使用命令安装（网络非常慢，不推荐）

在 Windows Terminal 或 PowerShell 中运行默认安装命令：

```powershell
wsl --install
```


该命令会自动安装 WSL 和默认 Ubuntu。它依赖在线下载，网络不稳定，因此容易失败。

### 第三步：首次启动 Ubuntu

安装完成后，重启电脑，打开 Ubuntu，按提示创建 Linux 用户名和密码。用户名不需要和 Windows 用户名相同；密码输入时通常不会显示字符，这是正常现象。



## 验证环境

进入 Ubuntu 后，按照下面的检查清单逐项验证环境。

### 1. 查看 Ubuntu 版本

```bash
cat /etc/os-release
```

当前默认版本通常为 `Ubuntu 24.04.1 LTS`。如果输出中能看到 `VERSION="24.04.1 LTS (Noble Numbat)"` 或类似的 `24.04` 信息，说明 Ubuntu 版本符合课程要求。

### 2. 确认基础命令可用

继续运行：

```bash
pwd
whoami
uname -a
bash --version
git --version
python3 --version
```

检查结果：

- `pwd` 能显示当前目录。
- `whoami` 能显示当前 Linux 用户名。
- `uname -a` 能显示 Linux 内核信息。
- `bash --version` 能显示 Bash 版本。
- `git --version` 能显示 Git 版本。
- `python3 --version` 能显示 Python 版本。

如果某条命令显示 `command not found`，先复制完整报错信息，课堂上再统一处理。

## 课堂检查清单

- 我能打开终端或 WSL。
- `pwd` 能显示当前目录。
- `git --version` 能显示 Git 版本。
- `python3 --version` 能显示 Python 版本。
