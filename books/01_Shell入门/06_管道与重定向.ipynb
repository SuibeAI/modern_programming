# 管道与重定向

Shell 的强大之处在于可以把小程序组合起来。组合的基础是标准输入、标准输出、标准错误、管道和重定向。

## 前提准备

先通过 `cd` 命令进入本章目录。下面的命令假设你已经在课程项目根目录。

```bash
cd books/01_Shell入门
pwd
ls -la
```

## 三类标准通道

- 标准输入（stdin）：程序读取数据的默认入口。
- 标准输出（stdout）：程序正常结果的默认出口。
- 标准错误（stderr）：程序错误信息的默认出口。

多数命令默认把结果打印到屏幕。重定向可以把结果写入文件，管道可以把结果交给下一条命令。

## 重定向

重定向用 `>` 和 `>>` 把命令输出写入文件，或用 `<` 从文件读入。这样可以把结果保存下来，便于查看、分享或作为下一步的输入。

### 覆盖输出 `>`


````python
%%bash
ls -la *.txt > output.txt
cat output.txt
````

`>` 会把标准输出写入指定文件。如果目标文件已经存在，会直接覆盖原内容，因此使用前要确认文件名。

这里的 `*` 是通配符（wildcard）。`*.txt` 表示匹配当前目录下任意以 `.txt` 结尾的文件，例如 `notes.txt`。

````python
%%bash
echo "=== 新增内容 ===" >> output.txt
echo "再次查看：" >> output.txt
ls -la *.txt >> output.txt
cat output.txt
````

`>>` 会把输出追加到文件末尾，而不是覆盖。这在记录日志或逐步积累结果时非常有用。

## 管道



管道 `|` 将前一个命令的标准输出连接到后一个命令的标准输入。可以理解为"然后交给"。

典型的管道模式是：**产生数据** → **筛选/转换** → **整理结果**。这样可以用简单命令的组合完成复杂任务。

````python
%%bash
# 例 1：在所有 Python 文件中搜索 TODO
# find: 查找所有 .py 文件
# xargs: 把结果逐个传给 grep
# grep -i: 不区分大小写地搜索
find ./ -name "*.py" | xargs grep -i "TODO"
````

````python
%%bash
# 例 2：查找所有 Python 文件中以 class 开头的行
find ./ -name "*.py" | xargs grep "^class"
````

````python
%%bash
# 例 3：统计目录大小，按从大到小排序，只显示前 10 项
# du -ah: 逐个文件显示大小
# sort -hr: 按数值从大到小排序（-h 理解 K/M/G 单位）
# head: 只显示前 10 行
du -ah . | sort -hr | head
````


### 小结

- `>` 覆盖输出，`>>` 追加输出，都是把数据保存到文件。
- 管道 `|` 把一个命令的输出连接到另一个命令的输入。
- 多个管道可以串联，形成处理流水线。
