# Prompt 与对话格式

预计阅读时间：10-15 分钟。

Prompt 是我们给模型的任务说明。写 prompt 的目标不是追求复杂，而是让模型清楚知道：要做什么、依据什么、按什么格式输出。

## 基本原则

写 prompt 时，先记住两点：

- 清晰具体：明确任务、范围、格式和限制；
- 给足上下文：如果任务依赖背景信息，就把背景写进 prompt。

清晰不等于短。很多时候，稍长但结构清楚的 prompt，比一句很短的要求更稳定。

## 简单 prompt

例如，我们想让模型介绍 Transformer，可以直接写：

```text
介绍 Transformer。
```

这个 prompt 能用，但比较模糊。模型不知道应该讲原理、应用、代码，还是论文背景；也不知道输出多长、用什么格式。

## 清晰 prompt

更好的写法是把任务、内容范围、输出格式和长度限制写清楚：

```text
请从模块层面介绍 Transformer，包括各个模块的输入输出维度。
采用 Markdown 格式返回，包括二级目录，不超过 200 字。
```

这个 prompt 比前一个更清楚，因为它明确了：

- 主题：Transformer；
- 角度：模块层面；
- 重点：输入输出维度；
- 格式：Markdown，包含二级目录；
- 长度：不超过 200 字。

## 使用分隔符

当 prompt 中包含较长文本、代码或日志时，建议用分隔符把输入内容包起来。例如：

````text
请总结下面的训练日志，只保留模型名、epoch、accuracy 和 loss。
日志如下：
```log
model=resnet18 epoch=1 accuracy=0.72 loss=0.84
model=resnet18 epoch=2 accuracy=0.78 loss=0.63
```
````

分隔符可以是三个反引号、XML 标签或明显的小标题。它的作用是告诉模型：哪些是任务说明，哪些是待处理内容。

## 对话格式

聊天模型通常使用 `messages` 表示对话。每条消息都有 `role` 和 `content`：

| role | 含义 |
| --- | --- |
| `system` | 设置模型行为和边界 |
| `user` | 用户输入的任务 |
| `assistant` | 模型已经给出的回复 |

例如：

`````python
messages = [
    {"role": "system", "content": "你是一个严谨、简洁的深度学习助教。"},
    {"role": "user", "content": "介绍 Transformer。"},
]
messages
`````

## 两个 prompt 对比

后面一节会用 API 运行这两个 prompt。这里先把它们保存成变量：

`````python
simple_prompt = "介绍 Transformer。"

clear_prompt = """请从模块层面介绍 Transformer，包括各个模块的输入输出维度。
采用 Markdown 格式返回，包括二级目录，不超过 200 字。"""

print(simple_prompt)
print()
print(clear_prompt)
`````

## 小结

一个实用 prompt 通常包含四部分：

- 任务：让模型做什么；
- 背景：模型需要参考哪些信息；
- 约束：长度、语气、范围和禁止事项；
- 格式：Markdown、JSON、列表或纯文本。

在项目中，prompt 往往会成为代码的一部分。越早把格式和边界写清楚，后续解析、测试和复用就越容易。
