# 语言模型基本原理

预计阅读时间：10-15 分钟。

语言模型的核心任务是：根据已有文本，预测后面最可能出现的内容。

## 从补全到对话

最简单的语言模型可以理解为文本补全器：

```text
输入：机器学习是一种
输出：让计算机从数据中学习规律的方法。
```

聊天模型在补全能力之上加入了对话格式、指令遵循和安全约束，因此看起来像是在回答问题。

## 生成过程

大模型不是一次性写完整段文字，而是一步一步生成 token。示意图如下：

![GPT-3 逐 token 生成示意图](figures/gpt3-generate-tokens-output.gif)

*GPT-3 逐 token 生成示意图。图片来源：[Jay Alammar, How GPT3 Works - Visualizations and Animations](https://jalammar.github.io/how-gpt3-works-visualizations-animations/)。*

具体而言，模型每一步只预测一个 token。这个 token 会被加入已有上下文，成为下一步预测的输入。

```mermaid
flowchart LR
    A[输入文本] --> B[切分为 token]
    B --> C[模型计算下一个 token 概率]
    C --> D[选择一个 token]
    D --> E[拼接到上下文]
    E --> C
```

每一步都会参考已有上下文，因此前面的输入会影响后面的输出。

## 架构直觉

![Transformer 模块示意图](figures/transformer-encoder-block.png)

*Transformer 模块示意图。图片来源：[Jay Alammar, The Illustrated GPT-2](https://jalammar.github.io/illustrated-gpt2/)。*

主流大模型通常基于 Transformer Decoder 架构。可以先把它理解为由若干相似模块反复堆叠而成，其中最关键的是两类计算：

- 自注意力机制：让模型在当前上下文中寻找相关信息，例如代词指代、前后约束和代码变量关系；
- MLP：对每个位置进一步变换表示，模型在训练中学到的大量语言规律、事实关联和代码模式，也会体现在这些参数中。

因此，大模型既会利用当前输入提供的上下文，也会利用训练中学到的知识和模式。这也是它能同时完成问答、总结、翻译和代码生成等任务的重要原因。

## 小结

从基本形式看，大模型是一个概率语言模型：它根据已有输入，预测下一个 token 出现的概率。

这种方式和人类显式写规则、做符号推理的方式不同。它看起来不太直观，但在大量数据和大规模参数的支持下，已经被证明非常有效。

也可以把大模型理解为一种压缩模型：训练过程把语言规律、事实关联和代码模式压缩进模型权重；生成回答时，模型再结合当前上下文，把这些模式重新组织成文本。

因此，本节最重要的结论是：大模型具备很强的上下文理解和文本生成能力。它不保证永远正确，但已经足够强大，值得我们系统学习和合理使用。
