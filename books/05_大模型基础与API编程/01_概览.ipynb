# 大模型基础与 API 编程概览

预计阅读时间：10 分钟。

本章介绍大模型的基本工作方式，以及如何通过 OpenAI 兼容 API 在 Python 程序中调用 GLM。重点不是训练模型，而是理解 prompt、对话消息、API 调用、结构化输出，以及如何把模型能力接入一个最小可用的应用。

## 学习目标

完成本章后，你应该能够：

- 解释大模型逐步生成 token 的基本过程；
- 理解 prompt、message、上下文和生成参数；
- 使用环境变量或 `.env` 保存 API Key；
- 使用 OpenAI 兼容 API 调用 GLM；
- 设计返回 JSON 的 prompt，并在业务代码中解析展示；
- 实现命令行版、Web 版和桌面版聊天应用；
- 理解多轮对话中历史消息的作用。

## API 编程的基本流程

```mermaid
flowchart LR
    A[准备 prompt] --> B[发送 API 请求]
    B --> C[模型生成回复]
    C --> D[解析输出]
    D --> E[进入业务代码]
```

API 调用本质上是一次网络请求。程序把 prompt、模型名称和参数发给服务端，服务端返回模型生成的内容。
